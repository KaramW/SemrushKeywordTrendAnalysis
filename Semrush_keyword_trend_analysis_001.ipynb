```python
# Step 1: Read the Excel file
excel_file_contnet= pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Semrush/androidpolice.com-organic.PositionChanges-us-20230711-2023-07-12T09_47_31Z.xlsx')
```

```python
df= pd.DataFrame(excel_file_contnet)
```

```python
df.head(10)
```

```python
# Convert the 'Trends' column into a list of integers
df['Trends'] = df['Trends'].apply(lambda x: [int(i) for i in x.split(',')])
```

```python
df.head(10)
```

```python
# Compute the average trend for each keyword
df['Average Trend'] = df['Trends'].apply(lambda x: sum(x) / len(x))
```

```python
df.head(10)
```

```python
# Convert 'Timestamp' to datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

```python
# Sort by 'Timestamp' to analyze trends over time
df = df.sort_values(by='Timestamp')
```

```python
df.head(10)
```

```python
# Get the top 10 URLs by average trend
top_urls = df.groupby('URL')['Average Trend'].mean().nlargest(100).index
```

```python
# Create a new DataFrame to store the keywords and search volume for rising trends
rising_trends_data = df[df['URL'].isin(top_urls)][['Keyword', 'Search Volume', 'Average Trend']]
```

```python
rising_trends_data
```

```python
# Remove duplicates
rising_trends_data = rising_trends_data.drop_duplicates()
```

```python
# Filter out the keywords that are less than 3 words long
rising_trends_data['num_words'] = rising_trends_data['Keyword'].apply(lambda x: len(x.split()))
```

```python
# Get only the keywords that are 3 words or longer
rising_trends_data_long = rising_trends_data[rising_trends_data['num_words'] >= 3]
```

```python
rising_trends_data_long
```

```python
len(rising_trends_data_long)
```

```python
# Drop the 'num_words' column as it's no longer needed
rising_trends_data_long = rising_trends_data_long.drop(columns=['num_words'])
```

```python
# Save the long keywords to an Excel file
rising_trends_data_long.to_excel('/content/drive/MyDrive/Colab Notebooks/Semrush/rising_trends_long_keywords.xlsx', index=False)
```
